```python
#inspired by Roettenbacher et al. 2018, Fig. 3

import numpy as np
import matplotlib.pyplot as plt

```

```python
x = np.linspace(0,1,1000)
y = np.full_like(x,1)
y[250:750] = np.cos((.5-x[250:750])*2*np.pi) +1
# y[300:800] += np.cos((.55-x[300:800])*2*np.pi)*0.5

plt.plot(x,y)
plt.ylim(0,2.5)
plt.xlim(0,1);
```
